```python
class ChatBot():
    def __init__(self, model='gpt-3.5-turbo'):
        self.model = model
        self.messages = []
        self.max_context_length = 4000

    def ask(self, question):
        if len(question) > self.max_context_length:
            question = question[:self.max_context_length]

        self.messages.append({
            'role': 'user', 
            'content': question
        })
        res = self.__ask__()
        return res
        
    def __ask__(self):
        my_messages = self.messages
        if len(self.messages) > self.max_context_length:
            my_messages = self.messages[:self.max_context_length]
            
        completion = openai.ChatCompletion.create(
            # model 지정
            model=self.model,
            messages=my_messages
        )
        response = completion.choices[0].message['content']
        self.messages.append({
            'role': 'assistant', 
            'content': response
        })
        return response
    
    def show_messages(self):
        return self.messages
    
    def clear(self):
        self.messages.clear()
```

```python
# model 파라미터에 다음의 모델 중 하나를 선택
#  'gpt-3.5-turbo',
#  'gpt-3.5-turbo-0301',
#  'gpt-4',
#  'gpt-4-0314'

# 인스턴스를 생성합니다.
chatbot = ChatBot(model='gpt-3.5-turbo')
```

```python
# 검색할 쿼리 설정 (예: 모든 문서 검색)
# query = {
#     "query": {
#         "multi_match": {
#             "query": "통계학과 인공지능",
#             "fields": ["*"]
#         }
#     }
# }

# query = {
#     "query": {
#         "match": {
#             "연구분야추가1": q
#         }
#     }
# }

q = f'메탄올'

query = {
    "query": {
        "bool": {
            "should": [
                {"match": {"연구정보추가1": q}},
                {"match": {"연구정보추가2": q}},
                {"match": {"연구정보추가3": q}},
                {"match": {"연구정보추가4": q}}
            ]
        }
    }
}
# 인덱스에서 데이터 검색
search_results = es.search(index='연구분야추가2', body=query, size = 20)

es_list = []
# 검색 결과 출력
for hit in search_results['hits']['hits']:
    source = hit['_source']
    score = hit['_score']
    # print({score}, {source['교수명']})
    print(f"Score: {score}, 교수명: {source['교수명']}, 연구정보: {source['연구분야통합']}")
    es_list.append(source)
```
