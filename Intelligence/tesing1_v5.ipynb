**1) get_description**: neo4j에 현재 저장된 objects와 predicates의 정보를 보여주는 함수. object와 predicate의 평균 갯수, 최소 갯수, 최대 갯수 등을 출력한다.

```python
cnt, info = db.get_description()
```

**2) get_object_list**: db에 저장되어 있는 object의 목록을 출력하는 함수

```python
db.get_object_list()
```

**4) get_predicate_list**: db에 저장되어 있는 predicate의 목록을 출력하는 함수

```python
db.get_predicate_list()
```
