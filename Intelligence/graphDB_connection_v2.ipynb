**2) get_object_list**: db에 저장되어 있는 object의 목록을 출력하는 함수

```python
db.get_object_list()
```

**3) get_object**: db에 노드로 저장되어 있는 object들을 출력하는 함수

> :param object:  The desired objects stored in the database.


```python
db.get_object(object = 'nan') 
```

**4) get_predicate_list**: db에 저장되어 있는 predicate의 목록을 출력하는 함수

```python
db.get_predicate_list()
```

**5) get_spo**: 특정 subject, object 혹은 predicate에 해당하는 spo를 추출하는 함수

> :param subject:    An argument that extracts spos with a specific subject.
>
> :param predicate:  An argument that extracts spos with a specific predicate.
>
> :param object:     An argument that extracts spos with a specific object.
>
> :param sp_link:    An argument that specifies how to extract specific subjects and predicates.
>
> :param po_link:    An argument that specifies how to extract specific predicates and objects.         
>
> :param so_link:    An argument that specifies how to extract specific subjects and objects.


```python
result = db.get_spo()
result
```

```python
for i, row in result.iterrows():
    print('video_id:', row['video_id'])
    print('video_path:', row['video_path'])
    print('captions:', row['captions'])
    print('spo:', row['spo'])
    print('frame:', row['frame'])
    
```
